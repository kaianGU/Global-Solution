```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/dataframe.csv', on_bad_lines='skip')
display(df)
```

```python
#removendo linhas e colunas vazias
df.dropna(axis=1, how='all', inplace=True)
df.dropna(axis=0, how='all', inplace=True)
```

# ***Estatísticas Básicas:***

```python
print(" Estatísticas Básicas: ")

def calcular_estatisticas(df, coluna):
    df[coluna] = pd.to_numeric(df[coluna], errors='coerce')
    estatisticas = {
        'Média': df[coluna].mean(),
        'Mediana': df[coluna].median(),
        'Moda': df[coluna].mode().values[0],
        'Desvio Padrão': df[coluna].std()
    }
    return estatisticas

resultados = {}
for coluna in ['Quantidade', 'Preço Unitário', 'Valor Total', 'Classificação do Produto (0 a 5)']:
    resultados[coluna] = calcular_estatisticas(df, coluna)

resultados_df = pd.DataFrame(resultados).T
print(resultados_df)

df.describe()
```

# ***Distribuição das Variáveis:***

```python
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.countplot(data=df, x='Categoria', palette='viridis')
plt.title('Distribuição por Categoria')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.countplot(data=df, x='Método de Pagamento', palette='viridis')
plt.title('Distribuição por Método de Pagamento')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
df["Categoria"] = df["Categoria"].astype(str)

# Verifique as primeiras linhas do DataFrame e as colunas
print(df.head())
print(df.columns)

# Assegure que as colunas são do tipo string, se necessário
df["Categoria"] = df["Categoria"].astype(str)
df["Método de Pagamento"] = df["Método de Pagamento"].astype(str)

# Configuração do estilo dos gráficos
sns.set(style="whitegrid")

# Distribuição da Categoria
plt.figure(figsize=(12, 6))
ax = sns.countplot(y="Categoria", data=df)
plt.title("Distribuição das Categorias")
plt.xlabel("Contagem")
plt.ylabel("Categoria")
ax.legend(title='Categoria', loc='upper right')
plt.show()

# Distribuição do Método de Pagamento
plt.figure(figsize=(12, 6))
ax = sns.countplot(y="Método de Pagamento", data=df)
plt.title("Distribuição dos Métodos de Pagamento")
plt.xlabel("Contagem")
plt.ylabel("Método de Pagamento")
ax.legend(title='Método de Pagamento', loc='upper right')
plt.show()

# Distribuição da Classificação do Produto
plt.figure(figsize=(10, 6))
ax = sns.histplot(df["Classificação do Produto (0 a 5)"], bins=20, kde=True)
plt.title("Distribuição da Classificação do Produto")
plt.xlabel("Classificação")
plt.ylabel("Frequência")
ax.legend(['Classificação'], title='Legenda', loc='upper right')
plt.show()

# Distribuição do Preço Unitário
plt.figure(figsize=(10, 6))
ax = sns.histplot(df["Preço Unitário"], bins=20, kde=True)
plt.title("Distribuição do Preço Unitário")
plt.xlabel("Preço Unitário")
plt.ylabel("Frequência")
ax.legend(['Preço Unitário'], title='Legenda', loc='upper right')
plt.show()

# Distribuição do Valor Total
plt.figure(figsize=(10, 6))
ax = sns.histplot(df["Valor Total"], bins=20, kde=True)
plt.title("Distribuição do Valor Total")
plt.xlabel("Valor Total")
plt.ylabel("Frequência")
ax.legend(['Valor Total'], title='Legenda', loc='upper right')
plt.show()
```

# ***Gráficos Univariados:***

```python
# Histograma da Classificação do Produto
plt.figure(figsize=(10, 6))
sns.histplot(df['Classificação do Produto (0 a 5)'], bins=20, kde=True, color='skyblue')
plt.title('Distribuição da Classificação do Produto')
plt.xlabel('Classificação do Produto')
plt.ylabel('Frequência')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Boxplot da Classificação do Produto por Categoria
plt.figure(figsize=(12, 6))
sns.boxplot(x='Categoria', y='Classificação do Produto (0 a 5)', data=df, palette='Set2')
plt.title('Boxplot da Classificação do Produto por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Classificação do Produto')
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Gráfico de Barras: Distribuição de Método de Pagamento
plt.figure(figsize=(12, 6))
sns.countplot(y='Método de Pagamento', data=df, palette='viridis')
plt.title('Distribuição do Método de Pagamento')
plt.xlabel('Contagem')
plt.ylabel('Método de Pagamento')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

# ***Gráficos Biváriados:***

```python
# Dispersão (Scatter Plot) entre Preço Unitário e Classificação do Produto
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Preço Unitário', y='Classificação do Produto (0 a 5)', data=df, hue='Categoria', palette='Set1', edgecolor='w', s=100)
plt.title('Relação entre Preço Unitário e Classificação do Produto')
plt.xlabel('Preço Unitário')
plt.ylabel('Classificação do Produto')
plt.legend(title='Categoria')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Gráfico de Barras Agrupadas: Média Geral por Categoria e Método de Pagamento
mean_ratings = df.groupby(['Categoria', 'Método de Pagamento'])['Classificação do Produto (0 a 5)'].mean().reset_index()
plt.figure(figsize=(14, 7))
sns.barplot(x='Categoria', y='Classificação do Produto (0 a 5)', hue='Método de Pagamento', data=mean_ratings, palette='tab10')
plt.title('Média da Classificação do Produto por Categoria e Método de Pagamento')
plt.xlabel('Categoria')
plt.ylabel('Média da Classificação do Produto')
plt.xticks(rotation=45)
plt.legend(title='Método de Pagamento')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

# ***Gráficos Multivariados***

```python
# Heatmap da Correlação entre Variáveis Numéricas
correlation_matrix = df[['Quantidade', 'Preço Unitário', 'Valor Total', 'Classificação do Produto (0 a 5)']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, vmin=-1, vmax=1)
plt.title('Heatmap de Correlação entre Variáveis Numéricas')
plt.show()
```
